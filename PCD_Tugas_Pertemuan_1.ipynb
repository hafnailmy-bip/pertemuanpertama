```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

# ==============================
# 1. MEMBACA CITRA
# ==============================
image_path = "Foto Tugas Pertemuan 1.jpeg"
image = cv2.imread(image_path)

if image is None:
    print("Gambar tidak ditemukan! Periksa path file.")
    exit()

# Konversi BGR ke RGB untuk ditampilkan dengan matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

print("Citra berhasil dibaca!")

# ==============================
# 2. REPRESENTASI MATRIKS & VEKTOR
# ==============================

print("\n===== REPRESENTASI MATRIKS (5x5 piksel pertama) =====")
print(image[0:5, 0:5])

vector = image.flatten()
print("\n===== REPRESENTASI VEKTOR (25 elemen pertama) =====")
print(vector[:25])

# ==============================
# 3. ANALISIS PARAMETER CITRA
# ==============================

height, width, channels = image.shape

print("\n===== ANALISIS PARAMETER =====")
print("Resolusi Spasial : {} x {}".format(width, height))

bit_depth = 8
total_bit_depth = bit_depth * channels
total_colors = 2 ** total_bit_depth

print("Bit Depth Total :", total_bit_depth, "bit")
print("Jumlah Intensitas/Warna Maksimum :", total_colors)

aspect_ratio = width / height
print("Aspect Ratio :", round(aspect_ratio, 2))

# Ukuran memori teoritis
memory_bytes = width * height * channels * bit_depth / 8
memory_kb = memory_bytes / 1024
memory_mb = memory_kb / 1024

print("Ukuran Memori Teoritis : {:.2f} MB".format(memory_mb))

# Simulasi: Resolusi 2x dan bit depth setengah
new_width = width * 2
new_height = height * 2
new_bit_depth = 4

new_memory = new_width * new_height * channels * new_bit_depth / 8
new_memory_mb = new_memory / 1024 / 1024

print("Memori jika resolusi 2x & bit depth setengah : {:.2f} MB".format(new_memory_mb))

# ==============================
# 4. MANIPULASI DASAR
# ==============================

# ---- Cropping ----
crop = image[100:400, 100:400]

# ---- Resizing (50%) ----
resize = cv2.resize(image, (width//2, height//2))

# ---- Rotasi 90 derajat ----
rotate = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

# ---- Flipping horizontal ----
flip = cv2.flip(image, 1)

# ==============================
# 5. MENAMPILKAN HASIL
# ==============================

plt.figure(figsize=(12,8))

plt.subplot(2,3,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(2,3,2)
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.title("Crop")
plt.axis("off")

plt.subplot(2,3,3)
plt.imshow(cv2.cvtColor(resize, cv2.COLOR_BGR2RGB))
plt.title("Resize 50%")
plt.axis("off")

plt.subplot(2,3,4)
plt.imshow(cv2.cvtColor(rotate, cv2.COLOR_BGR2RGB))
plt.title("Rotate 90")
plt.axis("off")

plt.subplot(2,3,5)
plt.imshow(cv2.cvtColor(flip, cv2.COLOR_BGR2RGB))
plt.title("Flip Horizontal")
plt.axis("off")

plt.tight_layout()
plt.show()

# ==============================
# 6. MENYIMPAN HASIL
# ==============================

cv2.imwrite("hasil_crop.jpg", crop)
cv2.imwrite("hasil_resize.jpg", resize)
cv2.imwrite("hasil_rotate.jpg", rotate)
cv2.imwrite("hasil_flip.jpg", flip)

print("\nSemua manipulasi selesai dan file hasil disimpan.")
```
